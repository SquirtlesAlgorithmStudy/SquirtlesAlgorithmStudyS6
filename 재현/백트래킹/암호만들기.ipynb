back-tracking : dfs에서 모든 경우를 탐색하지 않도록 조건을 걸어 이에 해당하면 이전 분기로 돌아가도록 하는 알고리즘  
  
Combination 구현과 비슷하게 진행  



```python
import sys

L, C = map(int, input().split(' '))

char = list(input().split(' '))
char.sort() # 사전순 출력이므로 배열먼저 진행
vowels = ['a', 'e', 'i', 'o', 'u'] #모음리스트
result = [0]*L #각 위치별로 인덱스 값이 들어감

def dfs(l, idx): # l : 분기의 깊이, idx : 문자열(char)의 index
    
    if l == L: #
        vo, co = 0, 0
        tmp = []
        for i in range(L):
            if char[result[i]] in vowels: #result에 저장된 index값으로 char의 문자를 뽑아서 tmp에 저장
                vo += 1
            else:
                co += 1
            tmp.append(char[result[i]])
            
        if vo >= 1 and co >= 2:
            print("".join(tmp))
        return
        
    else:
        for x in range(idx, C):
                
            result[l] = x #인덱스 값을 넣어줌 
            dfs(l+1, x+1)
                
                
            
dfs(0, 0)
```
