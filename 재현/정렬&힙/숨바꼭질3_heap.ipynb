## 아이디어  

heap : minheap으로 최소값을 반환할 수 있다.  
heap에 tuple을 입력값으로 넣었을 때 맨 앞의 값을 기준으로 최소 힙이 구성된다.  
앞의 값을 시간으로 넣어준다면 시간이 가장 작은 action(순간이동)을 먼저 heappop을 통해 반환하여 계산  


```python
from heapq import heappush, heappop
#heappop : 가장 작은값을 반환하고 제거
n, k = map(int, input().split())
sec = [0]*100001
visit = [False]*100001
visit[n] = True


heap = []
heappush(heap, (0, n))

while heap:
    time, loc = heappop(heap) # 현재 시간과 위치
    
    if loc == k:
        break
    for nx in [loc*2, loc-1, loc+1]:
        if 0 <= nx < 100001:
            if visit[nx] == False and nx == loc*2:
                heappush(heap, (time, nx)) # 순간이동은 time에 변경이 없다
                visit[nx] = True
                sec[nx] = time
                
            elif visit[nx] == False:
                heappush(heap, (time + 1, nx))
                visit[nx] = True
                sec[nx] = time + 1
                
print(sec[k])
```
