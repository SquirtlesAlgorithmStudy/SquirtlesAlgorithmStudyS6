```python
from collections import deque

n, k = map(int, input().split())

dx = [0, 1, -1]
sec = [-1] * 100001
visit = [False]*100001

queue = deque()
queue.append(n)
visit[n] = True
sec[n] = 0

while queue:

    x = queue.popleft()
    
    if x == k:
        break
        
    for i in range(3):
        if i == 0:
            nx = x*2
            if 0 <= nx <= 100000 and visit[nx] == False:
                queue.appendleft(nx)
                visit[nx] = True
                sec[nx] = sec[x]
        else:
            nx = x + dx[i]
            if 0 <= nx <= 100000 and visit[nx] == False:
                queue.append(nx)
                visit[nx] = True
                sec[nx] = sec[x] + 1
        

print(sec[k])
```
