## 아이디어 및 풀이  

1) 각 자릿수를 분해  
2) 문자별 자릿수의 합  
3) 정렬 후 큰 것부터 차례로 9부터 곱하여 더한다  


```python
n = int(input())

Num_dict = dict()
result = 0

for _ in range(n):
    num = input()
    digit = len(num) #해당 문자열의 자릿수 
    
    for d in range(digit): # 딕셔너리에 해당 문자가 없으면 자릿수에 해당하는 10 n제곱값을 넣어주고 
                           # 있으면(문자열 중복인 경우) 해당 자릿수에 해당하는 10 n제곱값을 추가로 더해줌
        Num_dict[num[d]] = Num_dict.get(num[d], 0) + 10**(digit - d - 1)
        
sorted_num_dict = dict(sorted(Num_dict.items(), key = lambda x: x[1], reverse = True)) #큰 수부터 곱해 더하기 위해 정렬

for i, y in enumerate(sorted_num_dict.values()): # 큰 자릿수에 9부터 매칭하여 곱하고 결과 도출
    result += (9 - i) * y

print(sorted_num_dict)
print(result)
```
